**Caderno de Exploração dos Dados do Datasus 2008 para frente**

```python
import pandas as pd
```

```python
dados = pd.read_csv("/content/A121613189_28_143_208_AnoMesProcessamento.csv", encoding="ISO-8859-1",
                    skiprows = 3, sep=";", skipfooter = 12,
                    thousands=".", decimal=",")
dados.head()
```

```python
dados.tail()
```

```python
pd.options.display.float_format = "{:.2f}".format
```

```python
dados.mean()
```

```python
dados["2008/Ago"]
```

```python
dados["2008/Ago"].mean()
```

```python
dados.plot(x = "Unidade da Federação", y = "2008/Ago", kind = "bar", figsize=(9,6))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x = "Unidade da Federação", y = "2008/Ago", kind = "bar", figsize=(9,6), color="g")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valores por Unidade da Federação")

#para sumir com as descrições acima do gráfico
plt.show()
```
