```python
dados[["2008/Ago","2008/Set"]].head()
```

```python
#média exclui as colunas que não tem todos os campos preenchidos, por esse motivo não vemos os dados entre 1992 a 2007/Jul
dados.mean()
```

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
colunas_usaveis
```

```python
#fazer um slice só das colunas úteis
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```python
usaveis = usaveis.set_index("Unidade da Federação")
```

```python
usaveis["2019/Ago"].head()
```

```python
usaveis.loc["12 Acre"]
```

```python
#droppar coluna "Total"
usaveis = usaveis.drop("Total", axis=1)
usaveis
```

```python
#pivotear o dataset
usaveis.T.plot(figsize=(10,6))
```

#Desafio: reposicionar a legenda do gráfico.

```python
ax = usaveis.T.plot(figsize=(15,7.5))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title('Gastos de Atendimentos Hospitalares por Estados do Brasil', fontsize = 20)
ax.set_ylabel('Valor (R$)', fontsize=14)
ax.set_xlabel('Mês', fontsize=14)
plt.show()
```

#Desafio: Plotar o gráfico de linha com apenas 5 estados de sua preferência

```python
ax = usaveis[:5].T.plot(figsize=(15,7.5))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title('Gastos de Atendimentos Hospitalares - Região Norte do Brasil', fontsize = 20)
ax.set_ylabel('Valor (R$)', fontsize=14)
ax.set_xlabel('Mês', fontsize=14)
plt.show()
```
